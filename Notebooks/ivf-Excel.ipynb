#### Helper functions

```python
allDays = np.array([3, 5, 7, 8, 9, 10, 11, 12, 13, 14, 15]) 

def daysInclusive (start = allDays[0], stop = allDays[-1]):
    return allDays[np.logical_and(allDays >= start,  allDays <= stop)]

def read_xl (file_name):
    xl_file = pd.ExcelFile( file_name )
    dfs = { sheet: xl_file.parse( sheet ) for sheet in xl_file.sheet_names }
    return dfs, xl_file.sheet_names
    
def getE2(df, days = allDays, addMissingColumns = False, interpolate = None):
    dayStrings = ['Day ' + str(day) + ' E2' for day in days]
    availableColumns =  [ d for d in dayStrings if d in df.columns] 
    print availableColumns
    df1 = df[availableColumns]
    if addMissingColumns:
        for day in dayStrings:
            if not day in df1:
                df1[day] = NaN
    if interpolate:
        nRows = len(df1.iloc[:])
        for i in range (nRows):
            if interpolate == True or interpolate == 'linear':
                df1.iloc[i] = df1.iloc[i].interpolate()
            elif interpolate == 'spline':
                df1.iloc[i] = df1.iloc[i].interpolate(method='spline')
            elif interpolate == 'polynomial':
                df1.iloc[i] = df1.iloc[i].interpolate(method='polynomial')
    return df1

def getL5DayString(day):
    return 'Day ' + str(day) + ' US (5 largest follicles)'
            
def getL5(df, days = allDays, addMissingColumns = False, interpolate = None):
    dayStrings = [getL5DayString(day) for day in days]
    availableColumns =  [ d for d in dayStrings if d in df.columns] 
    df1 = df[availableColumns]
    if addMissingColumns:
        for day in dayStrings:
            if not day in df1:
                df1[day] = NaN
    L5newCol = [float(str(d) + '.' + str(n)) for d in days for n in range (1, 6)]
    L5new = pd.DataFrame(np.zeros((len(df), len(L5newCol))), columns=L5newCol, dtype=str)
    
    #print L5new                          
    for day in days:
        s = getL5DayString(day)
        col = df1[s]
        for row in range(len(df1)):
            c = col[row]
            if c == None:
                continue 
            elif type(c) == object:
                continue
            elif type(c) == float or type(c) == np.float64:
                L5new.iloc[row][float(str(day)+'.1')] = c
            else:
                f = None
                try:
                    vf = c.split(',')
                    f = [float(f) for f in vf]
                    f = sort(f)
                    for n in range(1, len(f)):
                        L5new.iloc[row][float(str(day)+'.'+str(n))] = f[n]
                except Exception as ex:
                    print 'exception ', 'Day: ' + str(day), 'Row: ' + str(row), 'Type: ' + str(type(c)), ex, f
                    pass
                   
        
    if interpolate:
        nRows = len(L5new.iloc[:])
        for j in range (0, 5):
            colsel = L5newCol[j::5] # every 5th column
            print colsel
            for i in range (nRows):
                if interpolate == True or interpolate == 'linear':
                    print L5new.iloc[i][colsel]
                    L5new.iloc[i][colsel] = L5new.iloc[i][colsel].interpolate()
                elif interpolate == 'spline':
                    L5new.iloc[i] = L5new.iloc[i].interpolate(method='spline')
                elif interpolate == 'polynomial':
                    L5new.iloc[i] = L5new.iloc[i].interpolate(method='polynomial')
    return L5new

            
def getL5S(df, days = allDays, addMissingColumns = False, interpolate = None):
    dayStrings = [getL5DayString(day) for day in days]
    availableColumns =  [ d for d in dayStrings if d in df.columns] 
    df1 = df[availableColumns]
    if addMissingColumns:
        for day in dayStrings:
            if not day in df1:
                df1[day] = NaN
    L5newCol = [str(d) + '.' + str(n) for d in days for n in range (1, 6)]
    L5new = pd.DataFrame(np.zeros((len(df), len(L5newCol))), columns=L5newCol, dtype=str)
    
    #print L5new                          
    for day in days:
        s = getL5DayString(day)
        col = df1[s]
        for row in range(len(df1)):
            c = col[row]
            if c == None:
                continue 
            elif type(c) == object:
                continue
            elif type(c) == float or type(c) == np.float64:
                L5new.iloc[row][str(day)+'.1'] = c
            else:
                f = None
                try:
                    vf = c.split(',')
                    f = [float(f) for f in vf]
                    f = sort(f)
                    for n in range(1, len(f)):
                        L5new.iloc[row][str(day)+'.'+str(n)] = f[n]
                except Exception as ex:
                    print 'exception ', 'Day: ' + str(day), 'Row: ' + str(row), 'Type: ' + str(type(c)), ex, f
                    pass
                   
        
    if interpolate:
        nRows = len(L5new.iloc[:])
        for j in range (0, 5):
            colsel = L5newCol[j::5] # every 5th column
            print colsel
            for i in range (nRows):
                if interpolate == True or interpolate == 'linear':
                    print L5new.iloc[i][colsel]
                    L5new.iloc[i][colsel] = L5new.iloc[i][colsel].interpolate()
                elif interpolate == 'spline':
                    L5new.iloc[i] = L5new.iloc[i].interpolate(method='spline')
                elif interpolate == 'polynomial':
                    L5new.iloc[i] = L5new.iloc[i].interpolate(method='polynomial')
    return L5new
```
